def initialize_inventory(products):

````python
def initialize_inventory(products):
    return {product: 10 for product in products}  # Assume initial stock of 10 for each product.


def calculate_total_price(product):
    while True:
        try:
            price = float(input(f"Enter the price for {product}: "))
            if price < 0:
                raise ValueError("Price cannot be negative.")
            return price
        except ValueError as e:
            print(f"Error: {e}. Please enter a valid price.")


def get_customer_orders(products, inventory):
    customer_orders = {}
    while True:
        try:
            num_orders = int(input("How many different products are in the order? "))
            if num_orders <= 0:
                raise ValueError("The number of orders must be a positive number.")
            break
        except ValueError as e:
            print(f"Error: {e}. Please enter a valid number of orders.")
    
    for _ in range(num_orders):
        while True:
            product_name = input("Enter the product name: ").strip().lower()
            if product_name not in products:
                print("Error: Product not found. Please enter a valid product name.")
                continue

            try:
                quantity = int(input(f"Enter the quantity for {product_name}: "))
                if quantity <= 0:
                    raise ValueError("Quantity must be a positive number.")
                if inventory[product_name] < quantity:
                    print(f"Error: Only {inventory[product_name]} {product_name}s available in stock.")
                    continue
                customer_orders[product_name] = quantity
                break
            except ValueError as e:
                print(f"Error: {e}. Please enter a valid quantity.")

    return customer_orders


def calculate_order_statistics(customer_orders, products):
    total_price = 0
    total_quantity = 0
    for product, quantity in customer_orders.items():
        price = calculate_total_price(product)
        total_price += price * quantity
        total_quantity += quantity
    print(f"Total price: ${total_price:.2f}, Total quantity: {total_quantity}")


def update_inventory(customer_orders, inventory):
    for product, quantity in customer_orders.items():
        inventory[product] -= quantity


def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, stock in inventory.items():
        print(f"{product}: {stock} remaining")
````

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)

customer_orders = get_customer_orders(products, inventory)
calculate_order_statistics(customer_orders, products)
update_inventory(customer_orders, inventory)
print_updated_inventory(inventory)
````
